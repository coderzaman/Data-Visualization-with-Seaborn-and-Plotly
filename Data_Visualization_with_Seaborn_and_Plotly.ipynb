`sns.scatterplot(data=student, x="study_hours", y="test_score", hue="gender", style="class_level", size="Tshirt_size")`

---

### Core Parameters

* `data=student`
    This is the **dataset** you are using. It tells Seaborn to look inside the `student` DataFrame (or other data structure) to find all the columns mentioned next.

* `x="study_hours"`
    This sets the **horizontal x-axis**. The plot will position points based on the values in the "study_hours" column.

* `y="test_score"`
    This sets the **vertical y-axis**. The plot will position points based on the values in the "test_score" column.



---

### Semantic Mapping Parameters

These parameters add more dimensions to your 2D plot by changing the **appearance** (color, shape, size) of the points based on other columns.

* `hue="gender"`
    This parameter controls the **color** of the points. Each unique value in the "gender" column (e.g., 'Male', 'Female') will be assigned a different color. This is great for comparing groups.

* `style="class_level"`
    This controls the **marker style** (the shape) of the points. Each unique value in the "class_level" column (e.g., 'Freshman', 'Sophomore') will be assigned a different shape, like a circle, square, or triangle.

* `size="Tshirt_size"`
    This controls the **size** of the points. Each unique value in the "Tshirt_size" column (e.g., 'S', 'M', 'L', 'XL') will be assigned a different marker size. This is often used to show quantity or weight.

In summary, this single line of code creates a complex plot that shows the relationship between **study hours** and **test scores**, while also encoding information about **gender** (by color), **class level** (by shape), and **T-shirt size** (by point size) for every single student.
